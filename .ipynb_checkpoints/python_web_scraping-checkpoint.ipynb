## BeautifulSoup Í∞ùÏ≤¥ ÎßåÎì§Í∏∞
- BeautifulSoup ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º Ïù¥Ïö©ÌïòÏó¨ Ï†ïÏ†Å Ïõπ ÌéòÏù¥ÏßÄÏóêÏÑú Ï†ïÎ≥¥Î•º Ï∂îÏ∂úÌïúÎã§.
- ÌååÏã± : HTML ÏÜåÏä§ÏΩîÎìúÎ•º Ìï¥ÏÑùÌïòÎäî Í≥ºÏ†ï
BeautifulSoup Ìï®ÏàòÎäî Îß§Í∞úÎ≥ÄÏàòÎ°ú Ï†ÑÎã¨Î∞õÏùÄ html ÏÜåÏä§ÏΩîÎìúÎ•º Ìï¥ÏÑùÌï¥ÏÑú BeautifulSoup Í∞ùÏ≤¥Î•º ÏÉùÏÑ±ÌïúÎã§.
HTMLÏùÑ ÌååÏã±ÌïòÎäî Ï†ÅÏ†àÌïú Íµ¨Î¨∏ Ìï¥ÏÑùÍ∏∞Î•º Ìï®Íªò ÏûÖÎ†•Ìï¥Ïïº ÌïúÎã§.

print(type(soup)) => <class 'bs4.BeautifulSoup'>, Ï¶â bs4 ÎùºÏù¥Î∏åÎü¨Î¶¨ÏóêÏÑú Î∂àÎü¨Ïò® BeautifulSoup ÌÅ¥ÎûòÏä§ÎùºÎäî Í≤ÉÏùÑ ÌôïÏùºÌï† Ïàò ÏûàÎã§.




```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text

soup=BeautifulSoup(html_src,'html.parser')
print(type(soup))
print("\n")

print(soup.head) #htmlÏùò head Î∂ÄÎ∂Ñ Ï∂úÎ†•
print("\n")
print(soup.body) #htmlÏùò body Î∂ÄÎ∂Ñ Ï∂úÎ†•
print("\n")

print('title ÌÉúÍ∑∏ ÏöîÏÜå : ',soup.title) #Ïõπ Î∏åÎùºÏö∞Ï†Ä ÌÉ≠Ïóê ÌëúÏãúÎêòÎäî Î∂ÄÎ∂Ñ
print('title ÌÉúÍ∑∏ Ïù¥Î¶Ñ : ',soup.title.name) #html ÌÉúÍ∑∏Ïùò Ïù¥Î¶ÑÏùÑ Îî∞Î°ú Ï∂îÏ∂ú
print('title ÌÉúÍ∑∏ Î¨∏ÏûêÏó¥ : ',soup.title.string) #html ÌÉúÍ∑∏Î•º Ï†úÏô∏ÌïòÍ≥† Í∑∏ ÏïàÏóê ÌëúÏãúÎêòÎäî Î¨∏ÏûêÏó¥Îßå Ï∂úÎ†•
```

```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text

soup=BeautifulSoup(html_src,'html.parser') #HTML ÏÜåÏä§Î•º ÌååÏã±Ìï¥ÏÑú BeautifulSoup Í∞ùÏ≤¥Î•º ÏÉùÏÑ±ÌïúÎã§.

first_img=soup.find(name='img') #ÌÉúÍ∑∏Ïù¥Î¶Ñ(img)ÏùÑ name Îß§Í∞úÎ≥ÄÏàòÏóê Ï†ÄÏû•ÌïúÎã§.
print(first_img)
print("\n")

target_img=soup.find(name='img',attrs={'alt':'Seoul-Metro-2004-20070722.jpg'})
# ÌÉúÍ∑∏Ïù¥Î¶Ñ(img)ÏùÑ nameÏóê Ï†ÄÏû•, find() find()Î©îÏÜåÎìúÏùò attrs Îß§Í∞úÎ≥ÄÏàòÏóê {'ÏÜçÏÑ±Ïù¥Î¶Ñ':'ÏÜçÏÑ±Í∞í'}
# ÏßÄÏ†ïÌïú ÏÜçÏÑ±Í∞íÏùÑ Í∞ñÎäî ÌÉúÍ∑∏ Ï§ëÏóêÏÑú Í∞ÄÏû• Ï≤òÏùå ÎÇòÏò§Îäî ÌÉúÍ∑∏Î•º Ï∞æÎäîÎã§.
print(target_img)
```

## Ïõπ Î¨∏ÏÑúÏùò Í∑∏Î¶º Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏùÑ pcÏóê Ï†ÄÏû•ÌïòÍ∏∞
img ÌÉúÍ∑∏ ÏïàÏóê src ÏÜçÏÑ±ÏùÑ Îì§Ïó¨Îã§Î≥¥Î©¥ ÏÜåÏä§ÌååÏùº urlÏùÑ Î≥º Ïàò ÏûàÎäîÎç∞, Ïù¥Î•º Ïù¥Ïö©Ìï¥ÏÑú ÏßÅÏ†ë Îã§Ïö¥Î°úÎìú ÏãúÎèÑ
taget_img_src.get('ÏÜçÏÑ±Ïù¥Î¶Ñ') Î©îÏÜåÎìúÎ•º Ï†ÅÏö©ÌïòÎ©¥ ÏÜçÏÑ±Ïù¥ Í∞ñÎäî Í∞íÏùÑ Ï∂îÏ∂úÌï† Ïàò ÏûàÎã§.

>> ÌååÏùºÏùÑ Ïó¥Í≥† Ìï¥Îãπ Íµ¨Î¨∏Ïù¥ ÎÅùÎÇòÎ©¥ ÏûêÎèôÏúºÎ°ú Îã´ÌûàÍ≤å Ìï¥Ï£ºÎäî Í∏∞Îä•
>> with as Íµ¨Î¨∏ÏùÑ Îπ†Ï†∏ÎÇòÍ∞ÄÎ©¥ ÏûêÎèôÏúºÎ°ú close()Ìï®ÏàòÍ∞Ä Ìò∏Ï∂úÎêå
>> ÎÑ§Ìä∏ÏõåÌÅ¨ Ïä§Ìä∏Î¶ºÏùÑ Îã§Î£®Îäî ÏÜåÏºì ÌîÑÎ°úÍ∑∏ÎûòÎ∞çÍ≥º Í∞ôÏùÄ Í≥≥ÏóêÏÑúÎèÑ ÌôúÏö©ÌïúÎã§.

1. wb : Î∞îÏù¥ÎÑàÎ¶¨ Î™®ÎìúÏóêÏÑú Ïì∞Í∏∞ ÏûëÏóÖÌïòÎäî ÎèôÏïà ÌååÏùºÏùÄ Ïó¥Î†§Ï†∏ÏûàÎã§ (w+b)

```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text

soup=BeautifulSoup(html_src,'html.parser')

target_img=soup.find(name='img',attrs={'alt':'Seoul-Metro-2004-20070722.jpg'})
print('HTML ÏöîÏÜå : ', target_img)
print("\n")

target_img_src=target_img.get('src')
#taget_img_src.get('ÏÜçÏÑ±Ïù¥Î¶Ñ') Î©îÏÜåÎìúÎ•º Ï†ÅÏö©ÌïòÎ©¥ ÏÜçÏÑ±Ïù¥ Í∞ñÎäî Í∞íÏùÑ Ï∂îÏ∂úÌï† Ïàò ÏûàÎã§.
print('Ïù¥ÎØ∏ÏßÄ ÌååÏùº Í≤ΩÎ°ú : ', target_img_src)
print("\n")

target_img_resp=requests.get('http:'+target_img_src)
out_file_path="C:/Users/0206d/python_practice/output/download_image.jpg" #Ïù¥ÎØ∏ÏßÄ ÌååÏùº Ï†ÄÏû•Ìï† Í≤ΩÎ°úÎ•º ÏßÄÏ†ïÌï¥ÏÑú Î≥ÄÏàòÏóê Ï†ÄÏû•

with open(out_file_path,'wb') as out_file:
    out_file.write(target_img_resp.content) #Ïù¥ÎØ∏ÏßÄ ÌååÏùºÏù¥ Î∞îÏù¥ÎÑàÎ¶¨ ÌòïÌÉúÎ°ú Ï†ÄÏû•ÎêòÏñ¥ ÏûàÎã§.
    #write() Î™ÖÎ†πÏúºÎ°ú Ï†ÄÏû• ÏúÑÏπòÎ•º ÏßÄÏ†ïÌïòÏó¨ Ïô∏Î∂Ä ÌååÏùºÎ°ú Ï†ÄÏû•Ìï† Ïàò ÏûàÎã§.
    print('Ïù¥ÎØ∏ÏßÄ ÌååÏùºÎ°ú Ï†ÄÏû•ÌïòÏòÄÏäµÎãàÎã§')
    
```

## Ïõπ Î¨∏ÏÑúÏóê Ìè¨Ìï®Îêú Î™®Îì† ÌïòÏù¥ÌçºÎßÅÌÅ¨ Ï∂îÏ∂úÌïòÍ∏∞
- BeautifulSoup ÌÅ¥ÎûòÏä§Ïùò find_all() Î©îÏÜåÎìú ÏÇ¨Ïö©
- find_all("ÌÉúÍ∑∏Ïù¥Î¶Ñ") => Ïõπ Î¨∏ÏÑúÏóêÏÑú Ìï¥ÎãπÌïòÎäî Î™®Îì† ÌÉúÍ∑∏Î•º Ï∞æÏïÑÏÑú Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú Î¶¨ÌÑ¥ÌïúÎã§.
wiki_links=soup.find_all(name="a",href=re.complie("/wiki/"),limit=3)
>> a ÌÉúÍ∑∏Ïùò href ÏÜçÏÑ±Ïù¥ Ìè¨Ìï®ÌïòÎäî Î¨∏ÏûêÏó¥ÏùÑ Îî∞Î°ú ÏßÄÏ†ïÌïòÎ©¥ Ìï¥Îãπ Î¨∏ÏûêÏó¥Ïù¥ Ìè¨Ìï®Îêú aÌÉúÍ∑∏ÎßåÏùÑ Ï∞æÎäîÎã§.
>> Ï∂úÎ†• Í∏∏Ïù¥Î•º Ï†úÌïúÌïòÍ∏∞ ÏúÑÌï¥ limit Îß§Í∞úÎ≥ÄÏàòÎ•º 3ÏúºÎ°ú ÏÑ§Ï†ïÌïúÎã§.
>> Ï¶â /wiki/Î¨∏ÏûêÏó¥Ïù¥ ÎßÅÌÅ¨Ïóê Ìè¨Ìï®ÎêòÏñ¥ ÏûàÎäî a ÌÉúÍ∑∏Î•º 3Í∞ú Ï∞æÏïÑÏÑú Î≥ÄÏàòÏóê Ï†ÄÏû•ÌïúÎã§.

```python
import requests,re
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text
soup=BeautifulSoup(html_src,'html.parser')

links=soup.find_all("a")
print("ÌïòÏù¥ÌçºÎßÅÌÅ¨ Í∞úÏàò : ",len(links))
print("\n")
print("Ï≤´ 3Í∞úÏùò ÏõêÏÜå : ", links[:3])
print("\n")

wiki_links=soup.find_all(name="a",href=re.compile("/wiki/"),limit=3)
print("/wiki/ Î¨∏ÏûêÏó¥Ïù¥ Ìè¨Ìï®Îêú ÌïòÏù¥ÌçºÎßÅÌÅ¨ : ", wiki_links)
print("\n")

external_links=soup.find_all(name="a",attrs={"class":"external text"},limit=3)
print("class ÏÜçÏÑ±ÏúºÎ°ú Ï∂îÏ∂úÌïú ÌïòÏù¥ÌçºÎßÅÌÅ¨",external_links)
```

```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text
soup=BeautifulSoup(html_src,'html.parser')

subway_image=soup.select('#mw-content-text > div.mw-parser-output > table:nth-child(3) > tbody > tr:nth-child(1) > td > a > img')
print(subway_image)

print("\n")
print(subway_image[0])
print("\n")

subway_image2=soup.select('tr>td>a>img')
print(subway_image2[1]) #ÏÑúÏö∏ÏßÄÌïòÏ≤† Ïù¥ÎØ∏ÏßÄÎäî 2Î≤àÏß∏ ÏõêÏÜåÎãàÍπå Ïù∏Îç±Ïä§Î•º 1Î°ú ÏÑ§Ï†ïÌï®
```

## CSS Selector ÌôúÏö©ÌïòÍ∏∞2
- select()Î©îÏÑúÎìúÎäî find_all()Í≥º ÎπÑÏä∑ÌïòÎã§. Ï∞æÏúºÎ†§Îäî ÌÉúÍ∑∏ Ïù¥Î¶ÑÏùÑ Îß§Í∞úÎ≥ÄÏàòÎ°ú ÎÑòÍ≤®Ï£ºÎ©¥ Ïõπ Î¨∏ÏÑúÏóêÏÑú Ìï¥ÎãπÌïòÎäî Î™®Îì† ÌÉúÍ∑∏Î•º Ï∞æÏïÑÏÑú Î¶¨Ïä§Ìä∏ ÌòïÌÉúÎ°ú Î∞òÌôòÌï¥Ï§ÄÎã§.
- cssÏùò id ÏÑ†ÌÉùÏûêÎ•º ÌôúÏö©, id ÏÑ†ÌÉùÏûêÎäî Í≥†Ïú†Ìïú Í∞íÏùÑ Í∞ÄÏßÄÍ∏∞ ÎïåÎ¨∏Ïóê Ìïú Í∞úÏùò Î¶¨Ïä§Ìä∏Îßå Ï∞æÏïÑÎÇ∏Îã§. 
>> #idÏÜçÏÑ±Í∞í
- ÌÅ¥ÎûòÏä§ ÏÑ†ÌÉùÏûêÎ•º ÏÇ¨Ïö©ÌïòÎ©¥ Í∞ôÏùÄ ÌÅ¥ÎûòÏä§Î•º Í∞ñÎäî Ïó¨Îü¨ ÌÉúÍ∑∏Îì§ÏùÑ ÎèôÏãúÏóê Ï∞æÏùÑ Ïàò ÏûàÎã§.
>> .classÏÜçÏÑ±Í∞í ÏúºÎ°ú ÌëúÌòÑÌïúÎã§. 



```python
import requests
from bs4 import BeautifulSoup

url="https://en.wikipedia.org/wiki/Seoul_Metropolitan_Subway"
resp=requests.get(url)
html_src=resp.text
soup=BeautifulSoup(html_src,'html.parser')

links=soup.select('a')
print(len(links))
print("\n")

print(links[:3]) #Ïä¨ÎùºÏù¥Ïã± Í∏∞Î≤ï
print("\n")

external_links=soup.select('a[class="external text"]')
print(external_links[:3])
print("\n")

id_selector=soup.select('#siteNotice')
print(id_selector)
print("\n")

id_selector2=soup.select('div#siteNotice')
print(id_selector2)
print("\n")

id_selector3=soup.select('p#siteNotice')
print(id_selector3)
print("\n")

class_selector=soup.select('.mw-headline')
print(class_selector)
print("\n")

class_selector2=soup.select('span.mw-headline')
print(class_selector2)
```
